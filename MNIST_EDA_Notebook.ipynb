# MNIST Digit Dataset - Exploratory Data Analysis (EDA)

## Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load MNIST Data

```python
data_path = './data/mnist_train.csv'
data = pd.read_csv(data_path)
labels = data['label']
images = data.drop('label', axis=1)
```

## Visualize Sample Digits

```python
plt.figure(figsize=(10, 8))
for i in range(16):
    plt.subplot(4, 4, i+1)
    img = images.iloc[i].values.reshape(28, 28)
    plt.imshow(img, cmap='gray')
    plt.title(f'Label: {labels.iloc[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## Distribution of Labels

```python
plt.figure(figsize=(10, 6))
labels.value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of MNIST Labels')
plt.xlabel('Digit')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```
